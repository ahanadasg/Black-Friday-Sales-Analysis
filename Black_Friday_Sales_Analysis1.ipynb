```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data into the DataFrame 'df'
df = pd.read_csv('train.csv')

# Create the countplot for the 'Gender' variable
sns.countplot(x='Gender', data=df)

# Add a title to the plot
plt.title('Gender Distribution')

# Show the plot
plt.show()
```

```python
# Distribution of Age
sns.histplot(df['Age'], bins=10)
plt.title('Age Distribution')
plt.show()
```

```python
# Distribution of Age
sns.histplot(df['Age'], bins=10)
plt.title('Age Distribution')
plt.show()
```

```python
# Box plot of Purchase by Gender
sns.boxplot(x='Gender', y='Purchase', data=df)
plt.title('Purchase by Gender')
plt.show()
```

```python
# Bar plot of Purchase by City_Category
sns.barplot(x='City_Category', y='Purchase', data=df)
plt.title('Purchase by City Category')
plt.show()
```

```python

# Box plot of Purchase by Gender
sns.boxplot(x='Gender', y='Purchase', data=df)
plt.title('Purchase by Gender')
plt.show()
```

```python
# Bar plot of Purchase by City_Category
sns.barplot(x='City_Category', y='Purchase', data=df)
plt.title('Purchase by City Category')
plt.show()
```

```python
# Top 10 products by sales
top_products = df.groupby('Product_Category_1')['Purchase'].sum().nlargest(10)
print(top_products)
```
